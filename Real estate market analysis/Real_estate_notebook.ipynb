## Предобработка данных
<a id="2"></a>

### Работа с пропусками

Выведем матрицу корреляции для определения зависимости между показателями. Если будет обнаружена взаимосвязь, то пропуски в некоторых данных возможно заменить на медианные с учетом категорий.

```python
data.corr()
```

**Столбец balcony**<br>
Предположим, что если человек не указал число балконов, то их нет. Поэтому заменим пропуски значением 0.

```python
data['balcony'] = data['balcony'].fillna(0)
```

**Столбец ceiling_height**<br>
Поскольку между высотой потолка и другими данными отсутствует сильная взаимосвязь (коэффициент корреляци не превышает 0,1), можно заменить отсутствующие значения медианными.

```python
ceiling_height_median = data['ceiling_height'].median()
data['ceiling_height'] = data['ceiling_height'].fillna(ceiling_height_median)
```

**Столбец floors_total**<br>
Существует корреляция между floors_total и floor, что логично, т.к. количество этажей в определенной степени зависит от этажа квартиры (floors_total не может превышать floor). Поэтому замена пропусков в floors_total медианными значениями без учета этажа floor может исказить данные. Заменим пропуски медианными значениями с учетом этажа квартиры.

```python
# строим сводную таблицу с медианными значениями floors_total для каждого floor
floors_group = data.pivot_table(index='floor', values='floors_total', aggfunc='median')
floors_group = floors_group.rename(columns={'floors_total':'Медианная этажность'})

def floors_total_replace_nans(df):
    
    '''
    Возвращает медианную этажность из сводной таблицы,
    если в датафрейме df встречается NaN в столбце floors_total.
    В ином случае оставляем значение floors_total. 
    '''
    
    floor_number = df['floor']
    if pd.isna(df['floors_total']):
        return floors_group.loc[floor_number, 'Медианная этажность']
    return df['floors_total']

data['floors_total'] = data.apply(floors_total_replace_nans, axis=1)
```

**Столбец living_area**<br>
Жилая площадь сильнее всего зависит от общей площади квартиры total_area. Определим медианное значение доли жилой площади в квартирах и заполним пропуски исходя из этой доли.

```python
# соотношение жилой и общей площади
data['living_area_share'] = round(data['living_area'] / data['total_area'], 2)
living_area_share_median = data['living_area_share'].median()
living_area_share_median
```

```python
def living_area_replace_nans(df):
    '''
    Возвращает значение жилой площади living_area, полученное путем 
    умножения медианной доли жилой площади на общую площадь квартиры,
    если в датафрейме df встречается NaN в столбце living_area.
    В ином случае оставляем значение living_area.
    '''
    total_area = df['total_area']
    if pd.isna(df['living_area']):
        return total_area * living_area_share_median
    return df['living_area']

data['living_area'] = data.apply(living_area_replace_nans, axis=1)

data = data.drop(columns='living_area_share')
```

**Столбец kitchen_area**<br>
Аналогично площадь кухни также зависит от общей площади квартиры. Определим медианное значение доли кухни и заполним пропуски исходя из этой доли.

```python
# соотношение площади кухни и общей площади
data['kitchen_area_share'] = round(data['kitchen_area'] / data['total_area'], 2)
kitchen_area_share_median = data['kitchen_area_share'].median()
kitchen_area_share_median
```

```python
def kitchen_area_replace_nans(df):
   
    total_area = df['total_area']
    if pd.isna(df['kitchen_area']):
        return total_area * kitchen_area_share_median
    return df['kitchen_area']

data['kitchen_area'] = data.apply(kitchen_area_replace_nans, axis=1)

data = data.drop(columns='kitchen_area_share')
```

**Столбец is_apartment**<br>
Информация о том, является ли квартира апартаментами, не играет существенной роли для проведения требуемого анализа. Оставим пропуски.

**Столбец locality_name**<br>
Количество пропусков в назвниях населенных пунктах незначительное, заменим пропуски на "данные отсутствуют".

```python
data['locality_name'] = data['locality_name'].fillna('Данные отсутствуют')
```

**Остальные столбцы**<br>
Для оставшихся столбцов нет подходящего значения на замену, оставим пропуски.

### Изменение типов данных

В столбце first_day_exposition содержатся данные о дате публикации в фороме ISO. Для удобства восприятия и дальнейшей работой с датами приведем столбец к типу данных datetime.

```python
data['first_day_exposition'] = pd.to_datetime(data['first_day_exposition'], format='%Y-%m-%dT%H:%M:%S')
```

Столбцы floors_total, balcony содержат целочисленную информацию, поэтому целесообразно изменить тип данных на int.

```python
data['floors_total'] = data['floors_total'].astype('int')
data['balcony'] = data['balcony'].astype('int')
```

## Расчёты и добавление результатов в таблицу
<a id="3"></a>

```python
# расчет цены квадратного метра
data['square_meter_price'] = (data['last_price'] / data['total_area']).astype('int')

# добавления дня недели, месяца и года публикации
data['weekday'] = data['first_day_exposition'].dt.weekday
data['month'] = data['first_day_exposition'].dt.month
data['year'] = data['first_day_exposition'].dt.year

# категоризация этажей (первый, последний, другой)
def floor_category(row):
    floors_total = row['floors_total']
    floor = row['floor']
    if floor == 1:
        return 'первый'
    if floor == floors_total:
        return 'последний'
    return 'другой'

data['floor_category'] = data.apply(floor_category, axis=1)

# соотношение жилой и общей площади
data['living_area_share'] = round(data['living_area'] / data['total_area'], 2)
# соотношение площади кухни и общей площади
data['kitchen_area_share'] = round(data['kitchen_area'] / data['total_area'], 2)
```

### Изучение параметров: площадь, цена, число комнат, высота потолков.

```python
def get_distribution(column, xlim_lower, xlim_upper):
    
    fig = plt.figure(figsize=(25,10))
    fig.suptitle(f'Распределение {column}', fontsize=20)
    ax1 = fig.add_subplot(2,2,1)
    ax2 = fig.add_subplot(2,2,2)
    ax3 = fig.add_subplot(2,2,3)
    ax3.set_xlim(xlim_lower,xlim_upper)
    ax4 = fig.add_subplot(2,2,4)
    ax4.set_xlim(xlim_lower,xlim_upper)
    fig.tight_layout()

    sns.histplot(x=column, data=data, ax=ax1)
    sns.boxplot(x=column, data=data, ax=ax2)
    sns.histplot(x=column, data=data, ax=ax3)
    sns.boxplot(x=column, data=data, ax=ax4)

    plt.show() 
    
    print(data[column].describe())
```

### Убираем редкие и выбивающиеся значения

```python
def is_rare(df):
    '''
    Возвращает True, если квартира соответствует
    одному из условий, определенных ранее
    '''
    return (df['total_area'] > 115 
            or df['last_price_mln'] > 20 
            or df['rooms'] > 6 
            or df['ceiling_height'] < 2.45 
            or df['ceiling_height'] > 4)

# добавляем столбец с признаком редкости квартиры
data['rare'] = data.apply(is_rare, axis=1) 
```

```python
filtered_share = len(data.query('rare == True')) / len(data)
print(f'Доля редких квартир = {filtered_share:.1%}')
```

```python
# формируем новый датасет data_filtered без редких квартир
data_filtered = data.query('rare == False')
len(data_filtered)
```
